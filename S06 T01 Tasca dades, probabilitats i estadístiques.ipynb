```python
import pandas as pd
import numpy as np
import seaborn as sns
import statistics as st
```

```python
# Importado un data set con los ganadores del Tour de Francia desde 1903 a 2019
df = pd.read_csv('tdf_winners.csv')
```

```python
df
```

**Detalle de los datos** 
- edition = Edición del tour de Francia.
- start_date = Inicio de la edición.
- winner_name = Nombre del ciclista ganador.
- winner_team = Equipo del ciclista ganador.
- distance = Distancia total (km).
- time_overall = Tiempo total en horas.
- time_margin = Diferencia entre el primero y segundo clasificado en horas.
- stage_wins = Estapas ganadas.
- stages_led = Etapas con el mallot amarillo (etapas lideradas).
- height = Altura del ciclista en metros.
- weight = Peso del ciclista en kg.
- age = Edad del ciclista.
- born = Fecha de nacimineto.
- died = Fercha de fallecimiento
- full_name = Nombre completo del ciclista.
- nickname = Alias del ciclista.
- birth_town = Ciudad de nacimiento del ciclista.
- birth_country = País de nacimiento del ciclista.
- nationality = Nacionalidad del ciclista.

```python
# Cálculo de la edad media de los ganadores del Tour de Francia
df['age'].mean()
```

```python
# Moda muestra los tres valores más repetidos. 26 es la moda.
df['age'].mode()
```

```python
# Mediana de edad
df['age'].median()
```

```python
# Desviación estandard de edad.
df['age'].std()
```

- **Exercici 2:**

Continuant amb les dades de tema esportiu, selecciona dos atributs i calcula'n la seva correlació.

```python
# Creado un nuevo df con las estapas ganadas y etapas lideradas por ciclista
df_stages = df[['winner_name','stage_wins','stages_led' ]]
df_stages
```

```python
# Correlación entre las etapas lideradas y etapas ganadas
corr = np.corrcoef(df_stages['stages_led'], df_stages['stage_wins'])
corr
```

 ### Nivell 2

- **Exercici 3:**

Continuant amb les dades de tema esportiu, calcula la correlació de tots els atributs entre sí i representa'ls en una matriu amb diferents colors d'intensitat.

```python
# datos para la matriz.
matrix = df [['distance', 'time_overall', 'time_margin', 'stage_wins', 'stages_led', 'age']]
matrix
```

```python
# Correlación 
corr_matrix=matrix.corr()
corr_matrix
```

```python
# Heatmap con distintas intensidades 
sns.heatmap(corr_matrix, annot=True, cmap='PuOr')
```
